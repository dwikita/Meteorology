```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import xarray as xr
from mpl_toolkits.basemap import Basemap
```

```python
data = xr.open_dataset('ERA5.nc')
data
```

```python
lat = data.latitude
lon = data.longitude
```

```python
t2m = data.t2m[1,:,:]
```

```python
#Plot Indonesian region

fig= plt.figure(figsize=(15,8))

m = Basemap(projection='cyl', llcrnrlon=90, llcrnrlat=-15, urcrnrlon=150, urcrnrlat=15, resolution='i')
m.drawcoastlines(1)
m.drawcountries()

parallels = np.arange(-15,15+0.25,5)
m.drawparallels(parallels, labels=[1,0,0,0], linewidth=0.5)
meridians = np.arange(90,150+0.25,10)
m.drawmeridians(meridians, labels=[0,0,0,1], linewidth=0.5)

cf = plt.contourf(lon,lat,t2m-273.15, cmap='jet',extend='both')
cb = plt.colorbar(cf, fraction=0.0235, pad=0.03 )
cb.set_label(' \u00b0K',  fontsize=15, rotation=0)

plt.show()
```

```python
#Plot the specific region

fig= plt.figure(figsize=(15,8))

m = Basemap(projection='cyl', llcrnrlon=105, llcrnrlat=-9.25, urcrnrlon=115, urcrnrlat=-5.25, resolution='i')
m.drawcoastlines(1)
m.drawcountries()

parallels = np.arange(-9.25,-5.25+0.25,1)
m.drawparallels(parallels, labels=[1,0,0,0], linewidth=0.5)
meridians = np.arange(105,115+0.25,1)
m.drawmeridians(meridians, labels=[0,0,0,1], linewidth=0.5)

cf = plt.contourf(lon[60:105],lat[80:100],t2m[80:100,60:105], cmap='jet',extend='both')
cb = plt.colorbar(cf, fraction=0.0235, pad=0.03 )
cb.set_label(' \u00b0K',  fontsize=15, rotation=0)

plt.show()
```
